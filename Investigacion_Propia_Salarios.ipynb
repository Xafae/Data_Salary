```python
import numpy as np
import pandas as pd

df_original = pd.read_csv(r"Henry_Modulo_1_Homework\Clase_4_Pandas\Salarios_DataSci_Dataset.csv")
df_original.head()
```

```python
df_copia = df_original[["Último salario mensual  o retiro BRUTO (en tu moneda local)", "Pagos en dólares", "Dedicación", "Trabajo de"]].copy()
lista_columnas = df_copia.columns.values
print(lista_columnas)
```

```python
df_copia.rename(columns={"Último salario mensual  o retiro BRUTO (en tu moneda local)": "Salario"}, inplace=True)
print(df_copia.columns.values)
df_copia.head()
```

```python
Filtro = (df_copia["Trabajo de"] == "Data Scientist")
df_copia = df_copia[Filtro]
df_copia
```

```python
index = (df_copia["Salario"] == ".")
df_copia = df_copia.drop(index = 3572)
```

```python
df_copia["Salario"] = df_copia.Salario.str.replace(",", ".")
```

```python
df_copia.sort_values("Salario", ascending = False)
```

```python
df_copia["Salario"] = pd.to_numeric(df_copia["Salario"], downcast="integer")
df_copia
```

```python
Filtro = (df_copia["Salario"] <= 1000)
df_copia[Filtro]
```

```python
df_copia.loc[df_copia["Salario"] <= 1000, ["Salario"]] *= 160
df_copia
```

```python
df_copia.sort_values("Salario", ascending = False)
```

```python
print(df_copia["Salario"].min())
print(df_copia["Salario"].max())
print(df_copia["Salario"].median())
print(df_copia["Salario"].mean() * 0.7)
```
